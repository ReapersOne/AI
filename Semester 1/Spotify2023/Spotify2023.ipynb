```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("spotify2023.csv", encoding='latin-1')
```

```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
sns.histplot(df['streams'], bins=50, kde=True)
```

```python
df.info()
```

```python
df['in_shazam_charts'] = df['in_shazam_charts'].str.replace(',', '')
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'])
```

```python
summary = {
    'Max': df['in_shazam_charts'].max(),
    'Q3': df['in_shazam_charts'].quantile(0.75),
    'Median': df['in_shazam_charts'].median(),
    'Q1': df['in_shazam_charts'].quantile(0.25),
    'Min': df['in_shazam_charts'].min()
}
summary
```

```python
df_winterspring = df[df['released_month'] <= 6]
df_summerfall = df[df['released_month'] > 6]

def overlayHisograms(bottom, top, x, bins):
    sns.histplot(data=bottom, x=x, bins=bins, color='orange')
    sns.histplot(data=top, x=x, bins=bins, color='blue')
    
```

```python
sns.histplot(data=df_winterspring, x="released_month", bins=5, color="orange")
sns.histplot(data=df_summerfall, x="released_month", bins=5, color="blue")
```
