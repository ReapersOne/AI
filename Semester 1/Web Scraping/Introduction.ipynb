```python
url = "https://www.scrapethissite.com/pages/simple/"
page = requests.get(url)
html_content = page.text

soup = BeautifulSoup(html_content, 'html.parser')
```

```python
h3s = soup.find_all('h3')
```

```python
h3s
```

```python
h3s = soup.find_all('h3')
countries = []
for h3 in h3s:
    country = h3.get_text()
    country = country.strip()
    countries.append(country)
```

```python
df = pd.DataFrame(countries, columns=["Country Name"])
```

```python
countries_with_a = df[df["Country Name"].str.contains("A", case=False)]
countries_with_a.to_csv("countries_with_a.csv", index=False)
```
