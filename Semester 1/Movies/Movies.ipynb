```python
def get_best_fit(data, x, y):
    '''
    Determines the slope (m) and intercept (b) of the best fit line
    for two variables, x and y.
    '''
    m = 0
    for count in range(100):
        m = m + 1
        b = 0

        # Calculate residuals (y = mx+b)
        res = 0
        for i in range(len(data)):
            pred_y = (m * data.loc[i][x]) + b
            act_y = data.loc[i][y]
            temp = act_y - pred_y 
            temp = temp ** 2
            res = res + temp
        print(res)
    return (m, b, res)
```

```python
def calculate_r_squared(x, y):
    mean_y = sum(y) / len(y)

    sst = sum((yi - mean_y) ** 2 for yi in y)

    n = len(x)
    numerator = sum((x[i] - (sum(x) / n)) * (y[i] - mean_y) for i in range(n))
    denominator = sum((x[i] - (sum(x) / n)) ** 2 for i in range(n))

    m = numerator / denominator  # Slope
    b = mean_y - m * (sum(x) / n)  # Intercept

    predictions = [m * xi + b for xi in x]
    ssr = sum((y[i] - predictions[i]) ** 2 for i in range(n))

    r_squared = 1 - (ssr / sst) if sst != 0 else 0

    return r_squared

x = df['budget']
y = df['popularity']

r_squared = calculate_r_squared(x, y)
print("Coefficient of Determination (R²):", r_squared)
print("The Percentage of R² is: ", r_squared*100)
```
