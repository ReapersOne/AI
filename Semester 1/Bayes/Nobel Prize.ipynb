```python
def scrape_page(URL):
    try:
        # Fetch the webpage
        response = requests.get(URL)
        response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {URL}: {e}")
        return None

    # Parse the HTML
    soup = BeautifulSoup(response.text, 'html.parser')
    str_soup = str(soup)

    # List of Ivy League schools
    ivy_schools = [
        "Princeton", "Harvard", "Columbia", "Yale",
        "Cornell", "Dartmouth", "Brown", "Pennsylvania"
    ]

    # Count occurrences of each school
    ivy_counts = {school: str_soup.count(school) for school in ivy_schools}

    # Calculate total count
    total_ivy_mentions = sum(ivy_counts.values())

    # Print results
    print(f"Total Ivy League mentions: {total_ivy_mentions}")
    for school, count in ivy_counts.items():
        print(f"{school}: {count}")

    # Return results for further use
    return {"total": total_ivy_mentions, "individual_counts": ivy_counts}

# Example usage
url = "https://en.wikipedia.org/wiki/List_of_Nobel_laureates_by_university_affiliation"  # Replace with the URL to scrape
results = scrape_page(url)
```

```python
def scrape_and_calculate_probability(URL):
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, 'html.parser')
    str_soup = str(soup)

    # List of Ivy League universities
    ivy_league = ["Princeton", "Harvard", "Columbia", "Yale", "Cornell", "Dartmouth", "Brown", "Pennsylvania"]

    # Dictionary to store counts
    ivy_count = {university: str_soup.count(university) for university in ivy_league}

    # Calculate total mentions
    total_mentions = sum(ivy_count.values())

    # Avoid division by zero
    if total_mentions == 0:
        print("No Ivy League mentions found.")
        return ivy_count, {}, 0

    # Calculate probabilities
    ivy_probabilities = {university: count / total_mentions for university, count in ivy_count.items()}

    print("Ivy League Mentions and Probabilities:")
    for university, count in ivy_count.items():
        probability = ivy_probabilities[university]
        print(f"{university}: Mentions = {count}, Probability = {probability:.2%}")

    print(f"Total Ivy League mentions: {total_mentions}")
    return ivy_count, ivy_probabilities, total_mentions
```

```python
scrape_and_calculate_probability('https://en.wikipedia.org/wiki/List_of_Nobel_laureates_by_university_affiliation')
```
